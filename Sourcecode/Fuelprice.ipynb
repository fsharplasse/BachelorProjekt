```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel("../Sourcecode/")
```

```python
xs = df['Date']
series1 = df['electricity'].astype(np.double)
s1mask = np.isfinite(series1)
series2 = df['Diesel'].astype(np.double)
s2mask = np.isfinite(series2)
series3 = df['gasoline'].astype(np.double)
s3mask = np.isfinite(series3)

plt.plot(xs[s1mask], series1[s1mask], linestyle='-', color = 'magenta', label = 'Electricity')
plt.plot(xs[s3mask], series3[s3mask], linestyle='-',color = 'blue', label = 'Petrol')
plt.plot(xs[s2mask], series2[s2mask], linestyle='-', color = 'cyan', label = 'Diesel')
plt.xlabel('Years')
plt.ylabel('Price')
plt.title('Yearly change in fuel and electricity prices')
plt.legend()
plt.show()
```
